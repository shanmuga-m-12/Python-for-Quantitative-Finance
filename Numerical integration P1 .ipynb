```python
import numpy as np
import yfinance as yf
import time
```

```python
S0 = 100
K = 140
r = 0.05
q = 0.01
sig = 0.3
T = 1
N = 2500
dS = 0.1

def logNormal(S, r, q, sig, S0, T):
    f = np.exp(-0.5*((np.log(S/S0) - (r - q - sig**2/2)* T ) / (sig * np.sqrt(T)))**2) / (sig * S * np.sqrt(2 * np.pi * T))
    return f
    

df = np.exp(-r * T)

# Call intrgral

SumC = 0
S = np.zeros((N, 1))
for j in range(N):
    S[j] = K + j * dS

tmp = logNormal(S, r, q, sig, S0, T)

for j in range(N):
    if j == 0 :
        wj = dS/2
    else:
        wj = dS
    SumC += (S[j] - K) * tmp[j] * wj

call_price = df * SumC

# put intrgral

SumP = 0 
S = np.zeros((N,1))
eta = K/N
for j in range(N):
    S[j] = 0.1 + j * eta

tmp = logNormal(S, r, q, sig, S0, T)

for j in range(N):
    if j == 0 :
        wj = eta/2
    else:
        wj = eta
    SumP += (K - S[j]) * tmp[j] * wj
put_price = df * SumP
print('call price : ', call_price)
print('put price : ', put_price)
```
