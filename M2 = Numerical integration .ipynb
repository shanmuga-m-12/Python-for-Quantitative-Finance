```python
import numpy as np
```

```python
S0 = 100
K = 140
r = 0.05
q = 0.01
sig = 0.3
T = 1
N = 2500
dS = 0.1

def logNormal(S, r, q, sig, S0, T):
    f = np.exp(-0.5*((np.log(S/S0) - (r - q - sig**2/2)* T ) / (sig * np.sqrt(T)))**2) / (sig * S * np.sqrt(2 * np.pi * T))
    return f
    

df = np.exp(-r * T)

# Call intrgral

SumC = 0
S = np.zeros((N, 1))
for j in range(N):
    S[j] = K + j * dS

tmp = logNormal(S, r, q, sig, S0, T)

for j in range(N):
    if j == 0 :
        wj = dS/2
    else:
        wj = dS
    SumC += (S[j] - K) * tmp[j] * wj

call_price = df * SumC

# put intrgral

SumP = 0 
S = np.zeros((N,1))
eta = K/N
for j in range(N):
    S[j] = 0.1 + j * eta

tmp = logNormal(S, r, q, sig, S0, T)

for j in range(N):
    if j == 0 :
        wj = eta/2
    else:
        wj = eta
    SumP += (K - S[j]) * tmp[j] * wj
put_price = df * SumP
print('call price : ', call_price)
print('put price : ', put_price)
```

```python


S0 = 100
K = 140
r = 0.05
q = 0.01
sig = 0.3
T = 1
N = 10000



def logNormal(S, r, q, sig, S0, T):
    f = np.exp(-0.5*((np.log(S/S0) - (r - q - sig**2/2)* T ) / (sig * np.sqrt(T)))**2) / (sig * S * np.sqrt(2 * np.pi * T))
    return f
    
def trapzLogN(K, r, q, sig, S0, T, N):
    df = np.exp(-r * T)
    S = np.linspace( 0.01, 3*S0, N)
    pdf = logNormal(S, r, q, sig, S0, T)
    call_payoff = np.maximum(S - K, 0)
    put_payoff = np.maximum(K - S, 0)
    call_pric = df * np.trapz(call_payoff * pdf, S)
    put_pric = df * np.trapz(put_payoff * pdf, S)
    return call_pric, put_pric
    
call_pric, put_pric = trapzLogN(K, r, q, sig, S0, T, N)

print(f'Call price: {call_pric:.3f}')
print(f'Put price: {put_pric:.3f}')
```
